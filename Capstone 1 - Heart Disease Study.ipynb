```python
import pandas as pd

##Import pandas for DataTable functionality

heart = pd.read_csv('ProcessedHeart.csv')

##Import CSV

heart.columns = ['Age', 'Sex', 'Chest Pain Type','Resting Blood Pressure','Serum Cholesterol',
              'Fasting Blood Sugar > 120 mg/dl','Resting Electrocardiographic Results',
              'Maximum Heart Rate', 'Exercise Induced Angina','Old Peak','Slope Peak',
              'Major Vessels Colored','Thal','Angiographic Disease Status']
##Names of the columns in DataSet



def correlation_matrix(heart):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure(figsize =(15,15))
    ax1 = fig.add_subplot(1,1,1)   ### Dont get
    cmap = cm.get_cmap('autumn', 3)
    cax = ax1.imshow(heart.corr(), interpolation="nearest", cmap=cmap)
    ax1.grid(True)
    plt.title('Heart Disease Correlations')
    labels=['Age', 'Sex', 'C P T','R B ','S Chol',
              'FBS','Rest ECG',
              'Max HR', 'Ex Ang','Old Peak','Slope Peak',
              'M C V','Thal','A D S']
    ax1.set_xticklabels(labels,fontsize=15)
    ax1.set_yticklabels(labels,fontsize=15)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=[0,.25,.5,.75,.8,.85,.90,.95,1])
    plt.show()

correlation_matrix(heart)

##Make a Heatmap - Use to find most correlated areas for further Exploratory Data Analysis



```
